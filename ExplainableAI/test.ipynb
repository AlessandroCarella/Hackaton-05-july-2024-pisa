```python
def find_duplicates(dataset):
    duplicates = {}
    seen = {}

    for index, row in dataset.iterrows():
        row_tuple = tuple(row)
        if row_tuple in seen:
            if seen[row_tuple] in duplicates:
                duplicates[seen[row_tuple]].append(index)
            else:
                duplicates[seen[row_tuple]] = [index]
        else:
            seen[row_tuple] = index
    
    return duplicates
```

```python
duplicate_indices = find_duplicates(Xtest)
```
